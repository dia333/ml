```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv(r"C:\Users\hp\Downloads\covid.csv.csv")
data
```

```python
data=data.drop(['Time'],axis=1)
data
```

```python
data=data.drop(['ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
data                
                
```

```python
data=data.reset_index()
data=data.drop(['Sno','index'],axis=1)
data
```

```python

march_data=data.iloc[0:529]
march_data['Active']=march_data['Confirmed']-march_data['Deaths']-march_data['Cured']
march_data=march_data.groupby(['State/UnionTerritory'])[['Cured','Deaths','Confirmed','Active']].sum().reset_index()
march_data
```

```python
april_data=data.iloc[529:1478]
april_data=april_data.groupby(['State/UnionTerritory'])[['Cured','Deaths','Confirmed']].sum().reset_index()
april_data
```

```python
may_data=data.iloc[1478:2100]
may_data=may_data.groupby(['State/UnionTerritory'])[['Cured','Deaths','Confirmed']].sum().reset_index()
may_data
```

```python
grouped_data=data.groupby('State/UnionTerritory')
grouped_data.get_group('Kerala').sum()
```

```python
VISUALISING DATA:
```

```python
1.BARGRAPH SHOWING ACTIVE CASES BY END OF MARCH:
```

```python
plt.figure(figsize=(12,6))
plt.bar(march_data['State/UnionTerritory'],march_data['Active'])
plt.xticks(march_data['State/UnionTerritory'],rotation='vertical',size=10)
plt.xlabel('States/UnionTerritories',size='15')
plt.ylabel('Active cases',size='15')
plt.title('Active cases in month of march-statewise',size='20')
plt.show()
```

```python
2.GRAPH SHOWING DIFFERENCES BETWEEN DEATHS,CONFIRMED AND CURED IN MONTH OF APRIL
```

```python
plt.figure(figsize=(12,6))
plt.plot(april_data['State/UnionTerritory'], april_data['Deaths'],color='black')
plt.plot(april_data['State/UnionTerritory'], april_data['Confirmed'],color='red')
plt.plot(april_data['State/UnionTerritory'], april_data['Cured'], color='green')
plt.title('Deaths, Confirmed, Recovered in april-statewise',size='20')
plt.xticks(april_data['State/UnionTerritory'],rotation='vertical',size='8')
plt.xlabel('States/UnionTerritories',size='15')
plt.ylabel('Count',size='15')
plt.show()
```

```python
3.BARCHART COMPARING DIFFERENCES IN CONFIRMED,CURED,DIED CASES IN MONTH OF MAY:
```

```python
pos = list(range(len(may_data['Confirmed'])))
width = 0.25
fig, ax = plt.subplots(figsize=(20,10))
plt.bar(pos,may_data['Confirmed'], width, alpha=0.5, color='yellow')
plt.bar([p + width for p in pos], may_data['Deaths'], width, alpha=0.5, color='red')
plt.bar([p + width*2 for p in pos],may_data['Cured'], width, alpha=0.5, color='green')
ax.set_ylabel('Count',size='15')
ax.set_xlabel('States/UnionTerritories',size='15')
ax.set_title('comparison in month of MAY',size='20')
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels(may_data['State/UnionTerritory'],rotation='vertical',size='8')
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(may_data['Confirmed'] + may_data['Deaths'] + may_data['Cured'])] )
plt.legend(['Confirmed', 'Deaths', 'Cured'], loc='upper left')                          
plt.show()
```
